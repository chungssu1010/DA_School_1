### Groupby
- 데이터를 그룹별로 분할
1. split : 원본 데이터에서 분류하고자 하는 데이터를 나눠준다.
2. apply : 분류된 데이터(그룹별)에 함수 적용
3. combine : 함수 적용된 데이터를 다시 합쳐줌
- 기준 : 범주형 데이터(남과 여, 지역 별, 혈액형, 국적, 학년, mbti)

## 데이터프레임 병합

```python
data = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]
    }, index=['a', 'b', 'c', 'd']
)

data2 = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]},
    index=['e', 'f', 'g', 'h']
    )
```

```python
data
```

```python
data2
```

```python
add_data = pd.concat([data, data2])  # 두 개의 데이터를 합쳐줌, axis=1 도 사용가능
add_data
```

```python
add_data3 = pd.concat([data, data2], axis=1)  # 인덱스 없애야됨
add_data3
```

```python
data3 = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]
    }, index=['a', 'b', 'c', 'd']
)

data4 = pd.DataFrame({
    '도시': ['서울', '부산', '제주', '서울'],
    '연도': [2015, 2015, 2011, 2020],
    '전입신고 수': [10040, 22300, 6000, 12400],
    '세금': [1004000, 2230000, 600000, 1240000]},
    index=['a', 'b', 'c', 'd']
    )
```

```python
add_data2 = pd.concat([data3, data4], ignore_index=True)  # ignore_index=True : 인덱스 재배열
add_data2
```

### Join option

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                  'F': ['F2', 'F3', 'F6', 'F7']}, index=[2, 3, 6, 7])
```

```python
df4
```

```python
pd.concat([df1, df4], axis=1, join='outer')  # default = outer(합집합)
```

```python
pd.concat([df1, df4], axis=1, join='inner')  # inner : 교집합
```

```python
pd.concat([df1, df4], axis=1).reindex(df1.index)  # left join
```

```python
pd.concat([df1, df4], axis=1).reindex(df4.index)  # right join
```

```python
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                  'A': ['A0', 'A4', 'A2', 'A3'],
            'B': ['B0', 'B4', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                  'C': ['C0', 'C1', 'C2', 'C3'],
                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.concat([left, right], axis=1, join='inner')
```

```python
mi = pd.merge(left, right)  # default : inner
mi
```

```python
mo = pd.merge(left, right, how='outer')
mo
```

```python
m_key = pd.merge(left, right, on='key')  # 원하는 값
m_key
```

```python
m_l = pd.merge(left, right, on='key', how='left')
m_l
```

```python
m_r = pd.merge(left, right, on='key', how='right')
m_r
```
